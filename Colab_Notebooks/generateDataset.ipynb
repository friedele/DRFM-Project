```python
from __future__ import print_function
from google.colab import files
import os, io, zipfile, sys
import os, io, zipfile, sys
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential,Input,Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import pathlib
#from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU

# Load Modules
sys.path.append('/content/drive/MyDrive/Modules')
%run /content/drive/MyDrive/Modules/mathFunc.ipynb

# Dataset Parameters
dataset_path = '/content/drive/MyDrive/Images/Training' # The classes or folders under here will be used
image_size=(32,32)
batch_size=32
num_classes=2
epochs=32

# Show example of using a different notebook
#print (sqroot(25))
#print (square(5))
# print (cube(5))

data_dir = pathlib.Path(dataset_path)
image_count = len(list(data_dir.glob('*/*.png')))
print(image_count)

batch_size = 32
img_height = 180
img_width = 180

# Load and explore dataset
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_batches = train_datagen.flow_from_directory(
        dataset_path,
        target_size=image_size,
        batch_size=batch_size,
        class_mode='binary',  # Can set to categorical if we have subsets of folders
        subset='training')

validation_batches = train_datagen.flow_from_directory(
        dataset_path,
        target_size=image_size,
        batch_size=batch_size,
        class_mode='binary',
        subset='validation')

test_batches = train_datagen.flow_from_directory(
        dataset_path,
        target_size=image_size,
        batch_size=batch_size,
        class_mode='binary', 
        shuffle=True,
        subset='validation')

# Model Neural Network
drfm_model = Sequential()
drfm_model.add(Conv2D(32, kernel_size=(3, 3),activation='linear',input_shape=(64,64,1),padding='same'))
drfm_model.add(LeakyReLU(alpha=0.1))
drfm_model.add(MaxPooling2D((2, 2),padding='same'))
drfm_model.add(Conv2D(64, (3, 3), activation='linear',padding='same'))
drfm_model.add(LeakyReLU(alpha=0.1))
drfm_model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
drfm_model.add(Conv2D(128, (3, 3), activation='linear',padding='same'))
drfm_model.add(LeakyReLU(alpha=0.1))                  
drfm_model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
drfm_model.add(Flatten())
drfm_model.add(Dense(128, activation='linear'))
drfm_model.add(LeakyReLU(alpha=0.1))                  
drfm_model.add(Dense(num_classes, activation='softmax'))

# Compile the Model
#drfm_model.compile(loss=keras.losses.categorical_crossentropy,metrics=['accuracy'])
drfm_model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy()],
)
#drfm_model.summary()

# Train the Model
drfm_train = drfm_model.fit(train_batches, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=validation_batches)

def plotImages(images_arr):
    fig, axes = plt.subplots(1,5)
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()

imgs, labels = train_batches[1]
#plotImages(imgs)
print(labels[:5])
print('Number of test images: ', len(test_batches))
```
